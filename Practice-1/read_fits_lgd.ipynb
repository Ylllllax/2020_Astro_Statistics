```python
import astropy.io.fits as fits 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#读取fits文件，将HDU（header data unit）赋给hdu
hdu=fits.open('photosample.fit')

print(len(hdu))
#可以用len(hdu)看一下hdu有几个单元，不同的单元储存不同的信息，每个单元的header会解释同一单元下的data是什么。

#CDD图像一般储存在第一个unit里，如下：
data=hdu[0].data
header=hdu[0].header

#print(header)
#print(data)

fig=plt.figure(figsize=[12,6])
ax = fig.add_subplot(111)

ax.set_xlabel('X')
ax.set_ylabel('Y')
#plt.imshow,将二维矩阵转换为图像，注意data的维度，这里data是二维，所以不用删减维度。用data.shaoe查看维度
#cmap为colorbar的颜色条，可以搜索matplotlib colormap查看可以用到的颜色条，这里用的'Greys'是灰度条
im=ax.imshow(data,cmap='Reds',origin='low')
#显示colorbar
fig.colorbar(im, ax=ax)
#为colorbar设置范围
im.set_clim(0,np.percentile(data.flatten(),98))
plt.show()

#ps:对图像的处理，只需要对data这个二维矩阵处理，对矩阵的处理可以查看numpy
```
