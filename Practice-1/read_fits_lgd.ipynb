```python
import astropy.io.fits as fits 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#fitsHDUheader data unithdu
hdu=fits.open('photosample.fit')

print(len(hdu))
#len(hdu)hduheaderdata

#CDDunit
data=hdu[0].data
header=hdu[0].header

#print(header)
#print(data)

fig=plt.figure(figsize=[12,6])
ax = fig.add_subplot(111)

ax.set_xlabel('X')
ax.set_ylabel('Y')
#plt.imshow,datadatadata.shaoe
#cmapcolorbarmatplotlib colormap'Greys'
im=ax.imshow(data,cmap='Reds',origin='low')
#colorbar
fig.colorbar(im, ax=ax)
#colorbar
im.set_clim(0,np.percentile(data.flatten(),98))
plt.show()

#ps:datanumpy
```
